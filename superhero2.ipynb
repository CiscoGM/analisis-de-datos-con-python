```python
#5. hacer la prueba de bondad y ajuste con shapiro-wilk para la estatura
from scipy.stats import shapiro
from scipy.stats import kstest,norm

sample = df['Height'].sample(n=50)
print('shapiro')
stat, p_value = shapiro(sample)
print('stat', stat)
print('p-value', p_value)
if p_value > 0.05:
  print('Acepta H0: tienen distribuccion normal')
else:
    print('Acepto H1: no tiene distribución normal')

print()
```

```python
#6. hacer la prueba de bondad y ajuste con kolmogorov-smirnov para la estatura
sample = df['Height'].sample(n=50)
print('kstest')
stat, p_value = kstest(sample, norm.cdf)
print('stat', stat)
print('p-value', p_value)
if p_value > 0.05:
  print('Acepta H0: tienen distribuccion normal')
else:
    print('Acepto H1: no tiene distribución normal')

print()
```

```python
#7. hacer la prueba de bondad y ajuste con anderson para la estatura 
import numpy as np
from scipy import stats

muestra = df['Height'].sample(n=50)

result = stats.anderson(muestra, dist='norm')

print(f'Estadistico de Anderson: {result.statistic}')
print(f'valores Criticos: {result.critical_values}')
```

```python
# 8. comprobar el teroema del limite central la estatura
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings("ignore")
```

```python
print("population mean",marvel_height.mean())
print("population median",marvel_height.median())
print("population mode",marvel_height.mode())
print("population skew",marvel_height.skew())
print("population kurtosis",marvel_height.kurtosis())
print("population std",marvel_height.std())
print("population var",marvel_height.var())
```

```python
sns.distplot(marvel_height,hist=False,kde=True,color='blue')
plt.xlabel('population_marvel_height')
plt.ylabel('probabilty density')
```

```python
mean=[]

def calc_sample_mean(data,sample_size,no_sample_means):
    for i in range (no_sample_means):
        sample=data.sample(n=sample_size)
        no_sample_mean=sample.mean()
        mean.append(no_sample_mean)
    return (mean)
```

```python
mean_2=calc_sample_mean(data=marvel_height,sample_size=2,no_sample_means=500)

sns.distplot(mean_2,color='b')
plt.xlabel('sample base height(sample size=2)')
plt.ylabel('probability density')
```

```python
mean_3=calc_sample_mean(data=marvel_height,sample_size=5,no_sample_means=500)

sns.distplot(mean_3,color='b')
plt.xlabel('sample base height(sample size=5)')
plt.ylabel('probability density')
```
