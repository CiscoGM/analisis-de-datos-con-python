```python
import pandas as pd
```

```python
df=pd.read_csv("datasets/super_hero_info.csv",sep=',',header=0)#para las tabuladores 
df.head(20)
```

```python
 #Filtrar  los superheroes que su Publisher sea de Marvel y los guardan en un dataframe de nombre df_marvel
df_marvel=df[df.Publisher=="Marvel Comics"]
df_marvel
   
```

```python


df_marvel_bad = df_marvel[df_marvel.Alignment == 'bad']
df_marvel_bad
```

```python
#1 cual es el promedio de estatura para un super heroe malo hombre de Marvel Comics
df_marvel_bad.Height.mean()
```

```python
#2 cual es el minimo de estatura para un super heroe malo mujer de Marvel Comics
df_marvel_bad_women=df_marvel_bad[df_marvel.Gender=='Female']
df_marvel_bad_women.Height.min()

```

```python
#3 cual es el mediana y media de estatura para un super heroe bueno hombre de Marvel Comics
df_marvel_good = df_marvel[df_marvel.Alignment == 'good']
df_marvel_good

df_marvel_bad_man=df_marvel_good[df_marvel.Gender=='Male']
df_marvel_bad_man.Height.mean()

    
```

```python
df_marvel_bad_man.Height.median()
```

```python
#4 cual es el desviación estandar de estatura para un super heroe bueno mujer de Marvel Comics
df_marvel_good_women=df_marvel_good[df_marvel.Gender=='Female']
df_marvel_good_women.Height.std()
```

```python
#5. cual es el VARIANZA de estatura para un super heroe bueno mujer de Marvel Comics
df_marvel_good_women.Height.var()
```

```python
#6. cuales son los cuartiles 25%, 50% y 75% de estatura para un super heroe 
# bueno mujer de Marvel Comics

df_marvel_good_women.Height.quantile([.25,.50,.75])
```

```python
#7. cual es el sesgo de estatura para un super heroe bueno mujer de Marvel Comics
df_marvel_good_women.Height.skew()
```

```python
#84. cual es la kurtosis de estatura para un super heroe bueno mujer de Marvel Comics
df_marvel_good_women.Height.kurtosis()
```
